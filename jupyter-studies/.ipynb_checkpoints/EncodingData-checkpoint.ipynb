```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df[['sex','class','embark_town']].isna().sum()
```

```python
df = df.dropna(subset='embark_town')
```

```python
df[['sex','class','embark_town']].isna().sum()
```

```python
df
```

```python
# One-hot Encoder
```

```python
df['sex'].value_counts()
```

```python
df['embark_town'].value_counts()
```

```python
df['class'].value_counts()
```

```python
df_onehot = pd.get_dummies(df, columns=['sex','embark_town'], drop_first=True)
```

```python
df_onehot
```

```python
# label Encoder
```

```python
label_encoder = LabelEncoder()
```

```python
df_label = df.copy()
```

```python
df_label['sex_encoded'] = label_encoder.fit_transform(df_label['sex'])
```

```python
df_label
```

```python
df
```

```python
df.drop('sex_encoded', axis=1)
```

```python
# Ordinal Encoder
```

```python
df_ordinal = df.copy()
```

```python
ordinal_encoder = OrdinalEncoder()
```

```python
class_order = ['Third', 'Second', 'First']
ordinal_encoder = OrdinalEncoder(categories=[class_order])
```

```python
df_ordinal['class_encoded'] = ordinal_encoder.fit_transform(df_ordinal[['class']])
```

```python
df_ordinal
```

```python
fig, axes = plt.subplots(1,4,figsize=(15,5))
df['sex'].value_counts().plot(kind='bar',title='Original Categorical',ax=axes[0])
df_label['sex_encoded'].value_counts().plot(kind='bar',title='Label Encoded',ax=axes[1])
df_onehot['sex_male'].value_counts().plot(kind='bar',title='One-hot Encoded',ax=axes[2])
df_ordinal['class_encoded'].value_counts().plot(kind='bar',title='Ordinal Encoded',ax=axes[3])
```
