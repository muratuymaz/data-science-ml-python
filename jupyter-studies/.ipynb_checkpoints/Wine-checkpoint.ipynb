```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("13-WineQT.csv")
```

```python
data
```

```python
plt.figure(figsize=(10,8))
sns.set_style("whitegrid")
sns.scatterplot(data=data,x="fixed acidity",y="volatile acidity",hue='alcohol')
plt.xlabel('fixed acidity')
plt.ylabel('volatile acidity')
plt.title('alcohol by acidity')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.set_style("whitegrid")
sns.scatterplot(data=data,x="free sulfur dioxide",y="total sulfur dioxide",hue='alcohol')
plt.xlabel('free sulfur dioxide')
plt.ylabel('total sulfur dioxide')
plt.title('alcohol by sulfur dioxide')
plt.show()
```

```python
data
```

```python
plt.figure(figsize=(10,8))
sns.set_style("whitegrid")
sns.lineplot(data=data,x="alcohol",y="quality")
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Alcohol vs Quality')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(y='alcohol',x='citric acid',hue='quality',data=data)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(y='fixed acidity',x='alcohol',hue='quality',data=data)
plt.show()
```

```python
data1 = data.corr()
data1
```

```python
sns.barplot(data=data1,x='fixed acidity',y='alcohol',hue='quality')
plt.show()
```
