```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1**2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
```

```python
type(my_fig)
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1,data2,'g')
my_axes.plot(data2,data1,'b')
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1,data1+2,color = 'b',linewidth=1.0, marker = "o", markersize = 5)
new_axes.plot(data1,data1+3,color = 'y',linewidth=2.0, marker = "^")
plt.show()
```

```python
plt.scatter(data1,data2)
plt.show()
```

```python
plt.hist(data2,data1)
plt.show()
```
