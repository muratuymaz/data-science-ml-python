```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('athlete_events.csv')
```

```python
data.describe()
```

```python
data.head()
```

```python
data.isnull()
```

```python
data.info()
```

```python
plt.scatter("Height","Weight",data = data)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x="Height",y="Weight",hue="Sex",style="Medal",size="Age",data=data)
plt.xlabel('Height of atheletes')
plt.ylabel('Weight of atheletes')
plt.show()
```

```python
data.columns
```

```python
data['Medal'].unique()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height",y="Weight",data=data,hue="Sex")
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
```

```python
sns.set_style("dark")
sns.displot(x="Height",hue="Sex",data=data)
plt.ylabel("Frequency")
plt.xlabel('Height')
plt.show()
```

```python
sns.set_style("dark")
sns.displot(x="Height",hue="Sex",data=data,kind="kde")
plt.ylabel("Frequency")
plt.xlabel('Height')
plt.show()
```

```python
sns.barplot(x="Medal",y='Height',hue="Sex",data=data)
plt.title('Medals vs Height')
plt.show()
```

```python
sns.catplot(data=data,x='Medal',y='Height',hue='Sex',col='Season')
plt.show()
```
