```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Summary of Weather.csv')
df_location = pd.read_csv('Weather Station Locations.csv')
```

```python
df_clean = df.copy()
```

```python
df_clean
```

```python
group = ['DR', 'SPD', 'SND', 'FT', 'FB', 'FTI', 'ITH', 'PGT', 'TSHDSBRSGF', 'SD3', 'RHX', 'RHN', 'RVG', 'WTE', 'PoorWeather', 'WindGustSpd']
```

```python
df_clean.isnull().sum()
```

```python
df_clean = df_clean.drop(group, axis=1)
```

```python
df_clean.isnull().sum()
```

```python
group = ['Snowfall', 'PRCP', 'MAX', 'MIN', 'MEA', 'SNF']
```

```python
minRate = 0.90
rowSum = len(df_clean)
for column in group:
    nullData = df_clean[column].isnull().sum()
    if(nullData == 0):
        continue
    nullDataRate = nullData/rowSum
    if(nullDataRate>minRate):
        df_clean = df_clean.drop(column, axis=1, inplace=True)
    else:
        if(df_clean[column].dtype in ['int64', 'float64']):
            df_clean[column] = df_clean[column].fillna(df_clean[column].mean())
        else:
            df_clean[column] = df_clean[column].fillna(df_clean[column].mode()[0])
```

```python
df_clean.isnull().sum()
```

```python
df_clean.drop('YR', axis=1, inplace=True)
df_clean.drop('DA', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, ElasticNet, Ridge, LassoCV, ElasticNetCV, RidgeCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
scaler = StandardScaler()
regression = LinearRegression()
lasso = Lasso()
elasticnet = ElasticNet()
ridge = Ridge()
lassocv = LassoCV()
elasticnetcv = ElasticNetCV()
ridgecv = RidgeCV()
```

```python
# MaxTemp'i tahmin edecegimiz icin MaxTemp test icin kalan kolonlar train icin kullanilacak
X = df_clean.drop(columns='MaxTemp')
y = df_clean['MaxTemp']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=15, test_size=0.2)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
regression.fit(X_train_scaled, y_train)
```

```python
y_pred = regression.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print('mae:', mae)
print('mse:', mse)
print('score:', score)
plt.scatter(y_test, y_pred)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures()
```

```python
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
regression.fit(X_train_poly, y_train)
```

```python
y_pred = regression.predict(X_test_poly)
score = r2_score(y_test, y_pred)
print(score)
```

```python
poly = PolynomialFeatures(degree=4, include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
regression.fit(X_train_poly, y_train)
y_pred = regression.predict(X_test_poly)
score = r2_score(y_test, y_pred)
print(score)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print('mae:', mae)
print('mse:', mse)
print('score:', score)
plt.scatter(y_test, y_pred)
plt.show()
```
