```python
X = df[['Study Hours','Sleep Hours','Attendance Rate','Social Media Hours']]
y = df['Exam Score']

# y icin veriler son kolonda oldugu biliniyor ise su da kullanilabilir:
# X = df.iloc[:,;-1] -- son kolon haric diger kolonlari alir
# y = df.iloc[:,-1] -- sadece son kolonu alir
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
X_test[0]
```

```python
new_student = [[5, 7, 90, 4]]
new_student_scaled = scaler.transform(new_student)
regression.predict(new_student_scaled)
```

```python
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print('mse:', mse)
print('mae:', mae)
```

```python
score = r2_score(y_test, y_pred)
print('r2 score:', score)
```

```python
# adjusted score
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
