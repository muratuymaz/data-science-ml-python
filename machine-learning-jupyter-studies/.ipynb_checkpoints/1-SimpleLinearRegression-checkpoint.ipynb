```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('1-studyhours.csv')
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df,x='Study Hours',y='Exam Score')
plt.show()
```

```python
df.info()
```

```python
df.describe()
```

```python
# independent and dependent feature
X= df[['Study Hours']] # dataframe olarak daha kullanisli olur
y= df['Exam Score'] # series olarak daha kullanisli olur
```

```python
type(X)
```

```python
type(y)
```

```python
# test - train split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 15)
```

```python
y_test
```

```python
# standardize the data set
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
# model yavas calisiyorsa n_jobs= -1 yapabilirsin, bu butun islemci gucunu buraya kaydirmani saglar.
# regression = LinearRegression(n_jobs= -1)
```

```python
regression.fit(X_train, y_train)
```

```python
print('Coefficent:', regression.coef_)
print('Intercept:', regression.intercept_)
```

```python
# y = 76.91 + 16.18 x -- lineer regresyon denkleminin sabiti ve baskatsayisini buldu.
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regression.predict(X_train), 'r')
plt.show()
```

```python
# Standart scale islemi uygulandigi icin negatifler bulunuyor
```

```python
# x = 20, y = ?
```

```python
regression.predict([[20]])
# model burda standardization yapilmamis halini veriyor.
```

```python
scaler.transform([[20]])
# standardization islemi uygulanmis halinde kac girmemiz gerektigini buluyoruz.
```
